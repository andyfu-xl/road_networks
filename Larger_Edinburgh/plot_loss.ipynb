```python
import numpy as np
import matplotlib.pyplot as plt

# True values
true_values = np.array([1, 1])

# Generate a grid of predicted values
pred_range = np.linspace(0, 2, 100)
pred_x, pred_y = np.meshgrid(pred_range, pred_range)

# Calculate MSE for each combination of predicted values
mse_loss = ((pred_x - true_values[0]) ** 2 + (pred_y - true_values[1]) ** 2) / 2
# Calculate the directional loss
directional_loss = np.abs(np.arctan2(pred_y, pred_x) - np.arctan2(true_values[1], true_values[0]))

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
image1 = ax[0].contourf(pred_x, pred_y, mse_loss, levels=20, cmap='viridis')
ax[0].arrow(0, 0, 1, 1, 
            head_width=0.05, head_length=0.05, fc='grey', ec='grey')
ax[0].set_xlabel('Predicted x coordinate', size=14)
ax[0].set_ylabel('Predicted y coordinate', size=14)
ax[0].set_xlim(0, 2)
ax[0].set_ylim(0, 2)
ax[0].set_title('Heatmap of MSE Loss', size=16)
image2 = ax[1].contourf(pred_x, pred_y, mse_loss+directional_loss, levels=20, cmap='viridis')
ax[1].arrow(0, 0, 1, 1, 
            head_width=0.05, head_length=0.05, fc='grey', ec='grey')
fig.colorbar(mappable=image1)
fig.colorbar(mappable=image2)
ax[1].set_xlabel('Predicted x coordinate', size=14)
ax[1].set_ylabel('Predicted y coordinate', size=14)
ax[1].set_xlim(0, 2)
ax[1].set_ylim(0, 2)
ax[1].set_title('Heatmap of MSE Loss + Directional Loss', size=16)
plt.show()
# save the plot as pdf
fig.savefig('mse_vs_directional_loss.png', bbox_inches='tight', dpi=600)
```

```python
# Plotting the heatmap
plt.figure(figsize=(8, 6))
plt.contourf(pred_x, pred_y, mse_loss+directional_loss, levels=50, cmap='viridis')
plt.arrow(0, 0, 1, 1, 
          head_width=0.05, head_length=0.05, fc='grey', ec='grey')
plt.colorbar(label='MSE Loss')
plt.xlabel('Predicted Value 1')
plt.ylabel('Predicted Value 2')
plt.title('Heatmap of MSE Loss')
plt.show()
```
