```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dqn = np.load('results/test_rewards_dqn.npy')
gatten = np.load('results/test_rewards_gatten.npy')
gdqn = np.load('results/test_rewards_gdqn.npy')
sageatten = np.load('results/test_rewards_sageatten.npy')
```

```python
# box plot
mean_dqn = np.mean(dqn)
mean_gatten = np.mean(gatten)
mean_gdqn = np.mean(gdqn)
mean_sageatten = np.mean(sageatten)
fig = plt.figure(figsize =(10, 5))
plt.violinplot([dqn, gdqn, gatten, sageatten])
plt.scatter([1, 2, 3, 4], [mean_dqn, mean_gdqn, mean_gatten, mean_sageatten], color='red', s=100, label='Mean')
plt.xticks([1, 2, 3, 4], ['DQN', 'GCN', 'PA-GCN', 'PA-SAGE'], size=14)
plt.ylabel('Rewards', size=14)
plt.title('Acumulated Test Rewards', size=16)
plt.ylim(-5.5, 2.5)
plt.legend()
plt.savefig('violin_plot.pdf', format='pdf', dpi=600)
plt.show()
```

```python
test = np.load("results/test_rewards_SAGE_uncertain.npy")
import pandas as pd
df = pd.read_csv("results/reward_history_SAGE_uncertain.csv")
train = df['0'].values
```

```python
# violin plot
fig = plt.figure(figsize =(10, 5))
plt.violinplot([train, test])
plt.scatter([1, 2], [np.mean(train), np.mean(test)], color='red', s=100, label='Mean')
plt.xticks([1, 2], ['Train', 'Test'], size=14)
plt.ylabel('Rewards', size=14)
plt.title('Acumulated Rewards', size=16)
plt.ylim(-5.5, 2.5)
plt.legend()
plt.savefig('violin_plot.pdf', format='pdf', dpi=600)
plt.show()
```
